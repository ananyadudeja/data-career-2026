```python
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
```

Identify frequent categories

```python
#Which day is the busiest
print(df['day'].value_counts())
```

Compare group averages

```python
#Which time of the day has the highest average bill?
print(df.groupby('time')['total_bill'].mean())
```

Organise Results

```python
# Question: Which day has the highest average tip? (Sorted)
sorted_tips = df.groupby('day')['tip'].mean().sort_values(ascending=False)
print(sorted_tips)
```
